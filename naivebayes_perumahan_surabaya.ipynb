Klasifikasi Perumahan Surabaya dengan Menggunakan Algoritma Naive Bayes

```python
import pandas as pd
import numpy as np
from sklearn.naive_bayes import GaussianNB #Menunjukkan algoritma yang kita panggil adalah algoritma Naive Bayes
from sklearn import metrics
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report
from sklearn.metrics import ConfusionMatrixDisplay
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
# read csv file perumahan_surabaya
df = pd.read_csv('dataset_surabaya.csv')
df 
```

```python
# Menentukan nilai independent dan Dependent Variabel
x = df.iloc[:,:6].values
y = df.iloc[:,6].values
```

```python
x
```

```python
y
```

Membuat Model & Training

```python
nb_clf = GaussianNB()
nb_clf.fit(x,y)
```

Testing Model

```python
y_predict = nb_clf.predict(x)
y_predict
```

Evaluasi Model

```python
labels = ['Low', 'Medium', 'High']
akurasi = accuracy_score(y,y)
print('Tingkat Akurasi: %d persen'%(akurasi*100))
print(metrics.classification_report(y, y, labels=labels, target_names=labels))
ConfusionMatrixDisplay.from_predictions(y,y, labels=labels)
```
